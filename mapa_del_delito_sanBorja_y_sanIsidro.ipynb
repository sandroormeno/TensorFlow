```
#!pip install -q folium
import os
import folium
import numpy as np
import pandas as pd
import urllib.request
from folium.plugins import HeatMap
```

```
downloaded_data  = urllib.request.urlopen('https://raw.githubusercontent.com/sandroormeno/sim800L/master/imagenes/Mapa_de_Delitos.csv')
csv_data = pd.read_csv(downloaded_data, sep=',')
print (csv_data[:5])
```

```
delito = pd.DataFrame(csv_data)
arraydelito = np.array(delito) 
#pos =  np.array(arraydelito[169])
#print (len(arraydelito))
itemlist =[]

for j in range(2625):
    pos =  np.array(arraydelito[j])
    item = np.array([pos[1] , pos[0], 0.0001])
    itemlist.append(item)

#print (pos[0] , pos[1] , 0.01)
salida = np.array(itemlist).tolist()
print (salida[:10])
```

```
from folium.plugins import HeatMap

m = folium.Map([-12.1008821, -77.02110], tiles='stamentoner', zoom_start=14)

HeatMap(salida).add_to(m)

m
```
